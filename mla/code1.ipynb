```python
import csv;
with open("pvgl.csv",'v') as cfe:
    reader = csv.reader(cfe,delimiter = '.')
    train =[]
    for row in reader :
        train.append(row)
print(train)
```

```python
hypo = ['0']*6
row = 4
col = 6
```

```python
for i in range (row):
    if(train[i][6]=="yes"):
        for j in range (col):
            if(hypo[j]=="0"):
                hypo[j]=train[i][j]
            else:
                if(hypo[j]!=train[i][j]):
                    hypo[j]='?'
```
